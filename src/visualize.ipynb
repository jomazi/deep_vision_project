```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
from fcts.load_data import load_data
```

```python
data = load_data()
```

```python
img = Image.fromarray(data[20])
img.show()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(9, 6))

for i in range(12):
    img = Image.fromarray(data[i])
    ax = axes[int(i/4), i % 4]
    ax.imshow(img)
    ax.set_axis_off()

fig.suptitle("Icon Samples", fontsize=16, fontweight='bold')
plt.savefig('./plots/samples.png')
```
